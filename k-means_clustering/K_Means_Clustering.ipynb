# K-Means Clustering

### Import Dependencies

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

output = 'output'
try:
    os.mkdir(output)
except OSError as error:
    print(error)

np.random.seed(0)
```

### Create Demo Data

```python
X, y = make_blobs(n_samples=5000, centers=[[4, 4], [-2, -1], [2, -3], [1, 1]], cluster_std=0.9)

plt.scatter(X[:, 0], X[:, 1], marker='.')
#plt.savefig(output + '/k-means_1a.png')
```

### K-Means Clustering
* 4 clusters

```python
# Initialize KMeans clustering with specific parameters
k_means = KMeans(init="k-means++", n_clusters=4, n_init=12)

# Fit KMeans to the data
k_means.fit(X)

# Retrieve the labels assigned to each data point
k_means_labels = k_means.labels_

# Find the maximum value in the labels array
max_value = np.max(k_means_labels)
print('Maximum value of the array is', max_value)

# Calculate the length of the labels array
len(k_means_labels)

# Check the type of the KMeans object
type(k_means)

# Retrieve the cluster centers
k_means_cluster_centers = k_means.cluster_centers_

# Create a figure to plot the clustered data points
fig = plt.figure(figsize=(6, 4))

# Generate colors for each cluster
colors = plt.cm.Spectral(np.linspace(0, 1, len(set(k_means_labels))))

# Add a subplot to the figure
ax = fig.add_subplot(1, 1, 1)

# Plot each cluster's data points and cluster centers
for k, col in zip(range(len([[4, 4], [-2, -1], [2, -3], [1, 1]])), colors):
    # Select data points belonging to the current cluster
    my_members = (k_means_labels == k)

    # Retrieve the center of the current cluster
    cluster_center = k_means_cluster_centers[k]

    # Plot data points of the current cluster with specified color
    ax.plot(X[my_members, 0], X[my_members, 1], 'w', markerfacecolor=col, marker='.')

    # Plot the cluster center with a distinct marker
    ax.plot(cluster_center[0], cluster_center[1], 'o', markerfacecolor=col, markeredgecolor='k', markersize=6)

# Set title and remove ticks from the plot
ax.set_title('KMeans')
ax.set_xticks(())
ax.set_yticks(())

# Save the plot as an image
plt.savefig(output + '/k-means_1b.png')
```

### K-Means
* 3 clusters

```python
# Initialize KMeans clustering with different parameters
k_means3j = KMeans(init="k-means++", n_clusters=3, n_init=12)

# Fit KMeans to the data
k_means3j.fit(X)

# Create a figure to plot the clustered data points
fig = plt.figure(figsize=(6, 4))

# Generate colors for each cluster based on the labels
colors = plt.cm.Spectral(np.linspace(0, 1, len(set(k_means3j.labels_))))

# Add a subplot to the figure
ax = fig.add_subplot(1, 1, 1)

# Plot each cluster's data points and cluster centers
for k, col in zip(range(len([[4, 4], [-2, -1], [2, -3], [1, 1]])), colors):
    # Select data points belonging to the current cluster
    my_members = (k_means3j.labels_ == k)

    # Retrieve the center of the current cluster
    cluster_center = k_means3j.cluster_centers_[k]

    # Plot data points of the current cluster with specified color
    ax.plot(X[my_members, 0], X[my_members, 1], 'w', markerfacecolor=col, marker='.')

    # Plot the cluster center with a distinct marker
    ax.plot(cluster_center[0], cluster_center[1], 'o', markerfacecolor=col, markeredgecolor='k', markersize=6)

# Set title and remove ticks from the plot
ax.set_title('KMeans')
ax.set_xticks(())
ax.set_yticks(())

# Save the plot as an image
plt.savefig(output + '/k-means_1c.png')
```
